## Section 1: Variables and Basic Data Types

Python can store many kinds of values. Some common data types are:

- `int` → whole numbers (จำนวนเต็ม) like `5`, `12`, `100`
- `float` → decimal numbers (จำนวนทศนิยม) like `3.14`, `1.5`
- `str` → text or string (ข้อความ) like `"Alice"` or `"Hello"`
- `bool` → boolean (ค่าความจริง) which is `True` or `False`

You can store values in variables so you can use them later (เราสามารถเก็บค่าพวกนี้ไว้ในตัวแปรเพื่อเอาไปใช้ทีหลังได้)

### Example

```python
name = "Ploy"
age = 18
height = 1.62
is_student = True

print(name)
print(age)
print(height)
print(is_student)
```

## Section 2: Arithmetic Operators

Operators let us do calculations in Python.

Here are some important ones:

- `+` → addition
- `-` → subtraction
- `*` → multiplication
- `/` → division
- `//` → floor division (หารแบบปัดเศษทิ้ง เอาแค่จำนวนเต็ม)
- `%` → modulus (หารเอาเศษ)
- `**` → exponent (ยกกำลัง)

At first, we will use **static values**, which means we type the numbers directly into the code.

### Example

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
```

## Section 4: Getting Input from the User

Sometimes we do not want to hard-code values. We want the user to type something (บางทีเราก็อยากรับค่าจากผู้ใช้ครับ). 

We use `input()` for that.

### Example

```python
name = input("Enter your name: ")
print("Hello,", name)
```

**Important:** `input()` gives you text (`str`) by default. If you want to do math, you usually need to convert the input first  
(คำสั่ง input() จะคืนค่ามาเป็น string เสมอ ถ้าจะเอาไปคำนวณต้องแปลง data type ก่อนนะครับ)

For example:

```python
age = int(input("Enter your age: "))
height = float(input("Enter your height: "))
```
