## Section 1: What is a Conditional?

A conditional lets your program make a decision.

For example:

- If the user is older than 18, print `"Adult"`
- If a number is even, print `"Even"`
- If the password is wrong, print `"Access denied"`

This is what makes a program feel more interactive and useful.

### Basic Syntax

```python
if condition:
    # code to run if the condition is true
```

### Example

```python
age = 20

if age >= 18:
    print("You are an adult.")
```

If the condition is `True`, Python runs the indented code.
If the condition is `False`, Python skips it.


## Section 2: Comparison Operators

To make decisions, we compare values.

### Common Comparison Operators

| Operator | Meaning |
|---|---|
| `==` | equal to |
| `!=` | not equal to |
| `>` | greater than |
| `<` | less than |
| `>=` | greater than or equal to |
| `<=` | less than or equal to |

### Example

```python
score = 75

if score >= 50:
    print("Pass")
```

### Important Note

- `=` means **assign a value**
- `==` means **compare two values**

This is a very common beginner mistake, so pay attention to it.

## Section 3: Using `else`

Sometimes you want your program to do one thing if a condition is true, and a different thing if it is false.

That is what `else` is for.

### Syntax

```python
if condition:
    # runs if true
else:
    # runs if false
```

### Example

```python
age = 15

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

Exactly one branch will run.

## Section 4: Using `elif`

Sometimes there are more than two possible cases.

That is where `elif` helps.

### Syntax

```python
if condition_1:
    # runs if condition_1 is true
elif condition_2:
    # runs if condition_2 is true
else:
    # runs if none of the above are true
```

Python checks the conditions from top to bottom.
As soon as one condition is true, that block runs and the rest are skipped.

### Example

```python
score = 85

if score >= 80:
    print("A")
elif score >= 70:
    print("B")
elif score >= 60:
    print("C")
else:
    print("D")
```

## Section 5: Logical Operators

Sometimes one condition is not enough.
You may want to combine multiple conditions.

### Logical Operators

| Operator | Meaning |
|---|---|
| `and` | both conditions must be true |
| `or` | at least one condition must be true |
| `not` | reverses true/false |

### Examples

```python
age = 20
has_id = True

if age >= 18 and has_id:
    print("You may enter.")
```

```python
day = "Saturday"

if day == "Saturday" or day == "Sunday":
    print("Weekend")
```

```python
is_raining = False

if not is_raining:
    print("You can go outside.")
```

These are very useful in real programs.

## Section 6: Thinking Carefully About Conditions

A program may run without errors and still give the wrong answer.
That often happens when the conditions are written in the wrong order.

For example:

```python
score = 85

if score >= 60:
    print("C or above")
elif score >= 80:
    print("A")
```

This does **not** work the way you might expect, because `85 >= 60` is already true.
Python stops at the first true condition.

So when checking ranges, it is usually best to start from the highest condition and go downward.

This small idea matters a lot.
