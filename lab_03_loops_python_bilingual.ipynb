## Section 1: What is a Loop?

A **loop** repeats a block of code.

For example, imagine you want to print:

```python
Hello
Hello
Hello
Hello
Hello
```

You could write `print("Hello")` five times, but that is repetitive and not very efficient.

Instead, you can use a loop.

### Example

```python
for i in range(5):
    print("Hello")
```

This will print `"Hello"` 5 times.

### Important Idea

A loop is useful when:
- you want to repeat something many times
- you do not want to write the same code again and again
- you want your code to be shorter and cleaner

คำว่า **repeat** หรือ **iterate** เจอบ่อยมากเวลาเรียน loops  
ให้คิดง่าย ๆ ว่าคือการวนทำงานซ้ำเป็นรอบ ๆ



## Section 2: The `for` Loop

A `for` loop is often used when you know how many times you want to repeat something.

### Basic Syntax

```python
for variable in range(number):
    # code to repeat
```

### Example

```python
for i in range(3):
    print("Python")
```

### Output

```text
Python
Python
Python
```

### What is `i`?

In many beginner examples, `i` is just a variable name.
It changes each round of the loop.

In the example above:
- first round → `i` is `0`
- second round → `i` is `1`
- third round → `i` is `2`

ถึงแม้เราจะไม่ได้ใช้ `i` ใน `print()` ก็ยังใช้ได้  
เพราะมันเป็นแค่ตัวแปรที่ Python ใช้ไล่รอบให้เรา

## Section 3: Understanding `range()`

The `range()` function is used very often with `for` loops.

### Common Forms

```python
range(5)
```

This means:

```text
0, 1, 2, 3, 4
```

It starts at `0` and stops **before** `5`.

---

```python
range(2, 6)
```

This means:

```text
2, 3, 4, 5
```

It starts at `2` and stops **before** `6`.

---

```python
range(2, 11, 2)
```

This means:

```text
2, 4, 6, 8, 10
```

It starts at `2`, increases by `2`, and stops before `11`.

### Important Reminder

The stop value is **not included**.

อันนี้เป็นจุดที่นิสิตพลาดบ่อยมาก  
เช่น `range(5)` ไม่ได้แปลว่า 1 ถึง 5  
แต่มันคือ 0 ถึง 4

## Section 4: Using Loops with Calculations

Loops are not only for printing text.
They are also useful for calculations.

For example, you can use a loop to:
- find the sum of many numbers
- count how many times something happens
- repeat a calculation with changing values

### Example: Sum from 1 to 5

```python
total = 0

for i in range(1, 6):
    total = total + i

print(total)
```

### Output

```text
15
```

### What is happening here?

- `total` starts at `0`
- each round, we add the current value of `i`
- after the loop ends, `total` contains the final answer

คำว่า `total = total + i` ตอนแรกอาจดูงง  
ให้คิดว่าเป็นการเอาค่าเดิมของ `total` มาบวกเพิ่มทีละรอบ

## Section 5: The `while` Loop

A `while` loop repeats as long as a condition is `True`.

### Basic Syntax

```python
while condition:
    # code to repeat
```

### Example

```python
count = 1

while count <= 5:
    print(count)
    count = count + 1
```

### Output

```text
1
2
3
4
5
```

### Important Note

In a `while` loop, you must make sure the condition will eventually become `False`.

If not, the loop may run forever.
This is called an **infinite loop**.

ตัวอย่างเช่น ถ้าลืมเพิ่มค่า `count`  
loop ก็จะไม่จบ เพราะเงื่อนไขยังเป็นจริงตลอด

## Section 6: Loops with `if`

Loops and conditionals work very well together.

For example, you may want to:
- print only even numbers
- count only positive values
- check something inside each round

### Example

```python
for i in range(1, 11):
    if i % 2 == 0:
        print(i)
```

### Output

```text
2
4
6
8
10
```

แนวคิดคือ loop จะพาเราไล่ทีละค่า  
แล้ว `if` จะช่วยคัดเลือกว่าค่าไหนควรถูกใช้งานต่อ
