## Section 1: What is a String?

A string is text.

Examples:

```python
name = "Alice"
message = "Hello world"
```

Strings are written using quotes (`"` or `'`).

ภาษาไทย: string ก็คือ "ข้อความ" นั่นแหละ เช่น ชื่อ หรือประโยค

## Section 2: Combining Strings

You can combine strings using `+`

Example:

```python

second = "World"
print(first + " " + second)
```

Output:
```
Hello World
```

Useful methods:

