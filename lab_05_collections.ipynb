## Section 1: What is a Collection?

A collection stores multiple values in one variable.

For example, instead of writing:

```python
student1 = "Anna"
student2 = "Ben"
student3 = "Chris"
```

you can store them together:

```python
students = ["Anna", "Ben", "Chris"]
```

This is much easier to manage.

### Main Collection Types in This Lab

- `list` → ordered, changeable
- `tuple` → ordered, not changeable
- `set` → unordered, no duplicates
- `dict` → key-value pairs

แต่ละแบบไม่ได้มีไว้แทนกันทั้งหมด

## Section 2: Lists

### Example

## Section 4: Tuples

- allows duplicates

### When is a Tuple Useful?

A tuple is useful when the data should stay fixed, such as:

## Section 5: Lists vs Tuples

Both `list` and `tuple` store ordered data.

### Main Difference

- `list` → changeable
- `tuple` → not changeable

### Example

```python
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)
```

You can do this:

```python
my_list[0] = 99
```

But this will not work:

```python
my_tuple[0] = 99
```

### Quick Idea

Ask yourself:

- Will I need to edit the data later? → use `list`
- Should the data stay fixed? → use `tuple`

## Section 7: Dictionaries

A **dictionary** stores data as **key-value pairs**.

### Example

```python
student = {
    "name": "Alice",
    "age": 19,
    "major": "CS"
}
```

### Accessing Values

```python
print(student["name"])
print(student["age"])
```

### Important Properties of a Dictionary

- stores key-value pairs
- changeable
- keys should be unique

### Why Use a Dictionary?

A dictionary is useful when one item has multiple pieces of information.

For example:
- a student with name, age, and grade
- a product with title, price, and stock

ภาษาไทย: dict เหมาะกับข้อมูลที่มี "label" กำกับชัดเจน เช่น `name`, `price`, `stock`

## Section 8: Looping Through a Dictionary

You can loop through a dictionary in different ways.

### Loop through keys

```python
student = {"name": "Anna", "score": 90}

for key in student:
    print(key)
```

### Loop through values

```python
for value in student.values():
    print(value)
```

### Loop through both keys and values

```python
for key, value in student.items():
    print(key, value)
```

### Important Idea

This is very useful when you want to display or process structured data.

## Section 9: Quick Comparison

Here is a simple summary:

| Type | Ordered | Changeable | Duplicates | Main Use |
|---|---|---|---|---|
| `list` | Yes | Yes | Yes | general ordered items |
| `tuple` | Yes | No | Yes | fixed ordered items |
| `set` | No | Yes | No | unique items |
| `dict` | key-based | Yes | keys must be unique | labeled data |

### How to Think About Them

- `list` → "a sequence of items"
- `tuple` → "a fixed group of values"
- `set` → "a bag of unique values"
- `dict` → "data with labels"

ภาษาไทย: ถ้ายังสับสน ให้มองจาก "รูปแบบข้อมูล" ก่อน ไม่ต้องจำ definition แบบท่องอย่างเดียว
