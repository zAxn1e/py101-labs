## Section 1: What is a Function?

A **function** is a reusable block of code.

Instead of writing the same code again and again, you can put it inside a function and call it when needed.

### Example

```python
def greet():
    print("Hello!")
```

This creates a function named `greet`.

To run it, you must **call** it:

```python
greet()
```

### Output

```text
Hello!
```

### Important Idea

Defining a function does **not** run it yet.
It only creates the function.

You must call it to make it run.

ภาษาไทย:  
`def` คือการ "ประกาศ" ฟังก์ชัน  
ส่วนการใส่ `greet()` คือการ "เรียกใช้" ฟังก์ชัน

## Section 2: Parameters and Arguments

Functions can receive values.

### Example

```python
def greet(name):
    print("Hello,", name)
```

Here, `name` is a **parameter**.

When we call the function:

```python
greet("Mina")
```

`"Mina"` is an **argument**.

### Important Words

- **parameter** = variable inside the function definition
- **argument** = actual value sent into the function call

### Simple Pattern

```python
def function_name(parameter):
    # code
```

```python
function_name(argument)
```

ภาษาไทย:
- parameter = ตัวแปรที่ฟังก์ชัน "รอรับ"
- argument = ค่าจริงที่เราส่งเข้าไปตอนเรียกใช้

## Section 3: `print()` vs `return`

This is one of the most important ideas in this lab.

### `print()`

`print()` shows something on the screen.

### `return`

`return` sends a value back from the function so it can be used later.

### Example 1: Using `print()`

```python
def add_and_print(a, b):
    print(a + b)
```

### Example 2: Using `return`

```python
def add_and_return(a, b):
    return a + b
```

Now you can do this:

```python
result = add_and_return(3, 4)
print(result)
```

### Important Difference

- `print()` = display output
- `return` = give back a value

ภาษาไทย:
ถ้าใช้ `print()` อย่างเดียว ค่าจะถูกแสดงออกมา แต่เราเอาไปใช้ต่อไม่ได้ง่าย  
ถ้าใช้ `return` เราจะเก็บค่าไว้ในตัวแปร หรือส่งไปใช้ต่อกับฟังก์ชันอื่นได้

## Section 4: Function vs Method

This is another important concept.

### Function

A **function** is called directly by its name.

Examples:

```python
print("Hello")
len("Python")
```

### Method

A **method** belongs to an object and is called using a dot `.`

Examples:

```python
text = "Python"
print(text.upper())
```

```python
numbers = [1, 2]
numbers.append(3)
```

### Quick Rule

- no dot `.` → usually a function
- with dot `.` → usually a method

### Examples Side by Side

```python
text = "hello"

print(len(text))      # function
print(text.upper())   # method
```

ภาษาไทย:
- function เหมือนเครื่องมือกลาง เรียกใช้ได้เลย
- method คือความสามารถที่ "ติดอยู่กับ object นั้น"

เช่น:
- `len(text)` → function
- `text.upper()` → method ของ string
- `numbers.append(4)` → method ของ list

## Section 6: Functions with Loops

Functions can also contain loops.

### Example

```python
def print_numbers():
    for i in range(1, 6):
        print(i)
```

### Another Example

```python
def sum_list(numbers):
    total = 0
    for number in numbers:
        total += number
    return total
```

This is very useful because loops often represent repeated logic that we want to reuse.

## Section 7: Why Functions Matter

Functions help us:
- reduce repeated code
- organize logic
- make code easier to read
- make debugging easier
- reuse code in bigger programs

### Example Without Function

```python
score1 = 80
if score1 >= 50:
    print("Pass")
else:
    print("Fail")

score2 = 30
if score2 >= 50:
    print("Pass")
else:
    print("Fail")
```

### Example With Function

```python
def check_pass(score):
    if score >= 50:
        return "Pass"
    else:
        return "Fail"

print(check_pass(80))
print(check_pass(30))
```

ภาษาไทย:

## Section 8: Scope (Light Introduction)

A variable created inside a function is usually used only inside that function.

### Example

```python
def test():
    x = 10
    print(x)

test()
```

This works.

But this will usually not work:

```python
print(x)
```

because `x` was created inside the function.

### Simple Idea

- variable inside function → local variable
- variable outside function → outside/local to another scope

For now, the main thing to remember is:

> Variables created inside a function are mainly for that function.

ภาษาไทย:
ยังไม่ต้องลงลึกเรื่อง scope มาก  
แค่จำก่อนว่า ตัวแปรในฟังก์ชันมักไม่ควรไปหวังใช้ข้างนอกตรง ๆ
