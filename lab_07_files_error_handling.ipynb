## Section 1: Why Files Matter

So far, most of your programs have stored data only in variables.

Example:

```python
name = "Mina"
score = 90
```

The problem is that when the program ends, those values are gone.

A file lets you store data so it can still exist later.

This is useful for:
- notes
- scores
- logs
- saved settings
- simple records

### File Modes You Should Know

| Mode | Meaning |
|---|---|
| `"r"` | read file |
| `"w"` | write file (overwrite old content) |
| `"a"` | append to file |

### Important Idea

- `r` → read existing data
- `w` → start fresh and replace old data
- `a` → keep old data and add more

ภาษาไทย:
ถ้าอยาก "เก็บข้อมูลไว้" หลังโปรแกรมจบ ต้องเริ่มใช้ไฟล์
และต้องระวังด้วยว่า `w` จะทับข้อมูลเดิม

## Section 2: Overwrite vs Append

When working with files, it is very important to know whether your program should:

### Reminder

- `"w"` overwrites
- `"a"` appends

### Real-world thinking

Ask yourself:
- Do I want to reset the file? → use `"w"`
- Do I want to keep history and add more? → use `"a"`

This small choice matters a lot in practical programs.

## Section 4: What is Error Handling?

Programs do not always receive correct input.
Files may not exist.
Users may type text when a number is expected.

If we ignore these cases, the program may crash.

### Basic Structure

```python
try:
    # risky code
except:
    # what to do if an error happens
```

### Better Structure

Sometimes it is better to catch a specific error:

```python
try:
    number = int(input("Enter a number: "))
except ValueError:
    print("Invalid input")
```

### Important Idea

Error handling is not for hiding mistakes.
It is for helping the program fail more safely.

ภาษาไทย:
`try` คือ "ลองทำ"
`except` คือ "ถ้าพัง ให้มาทำตรงนี้แทน"
