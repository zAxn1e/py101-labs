## Section 3: Core Project Requirements

Your mini project **must** include all of the following.

### Functional Requirements

Your program must have:

1. **a clear purpose**
   - the user should be able to explain what the program is for

2. **a menu system or a clear repeated user flow**
   - the user should be able to do more than one action or interact more than once

3. **at least 1 loop**
   - for repeated interaction or repeated processing

4. **at least 1 conditional structure**
   - such as `if`, `elif`, `else`

5. **at least 3 functions**
   - your code should not be written as one giant block

6. **at least 1 collection**
   - such as `list`, `tuple`, `set`, or `dict`

7. **file reading and/or file writing**
   - your project should save data, load data, or both

8. **basic error handling**
   - use `try/except` where it makes sense

### Minimum Standard

The program should feel like one complete small system, not just many unrelated code parts placed together.



## Section 4: Technical Expectations

Your project should also meet these technical expectations:

- use meaningful variable names
- use meaningful function names
- keep code readable
- avoid repeating the same code many times
- print clear output for the user
- handle at least one realistic error case
- test more than one user input path

### Practical Reminder

A project is not only about passing the normal case.
It should also behave reasonably when:
- the user types the wrong kind of input
- a file does not exist yet
- the user enters unexpected values
